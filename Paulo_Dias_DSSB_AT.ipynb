### Resolução

1. Dentre as requisições HTTP o Spring Boot possui suporte para todas as principais sendo elas:

    * GET - Utilizado para requerir registros da API, tendo como base um identificador único (Eg.: ID, CPF, ETC);
    * POST - Utilizado para inserir registros na API, se utiliza do formato base do banco de dados, em formato JSON (Eg.: {"nome": "Marcos, "CPF": "123.456.789-10", ...}}
    * PUT - Possui utilização para atualização dos dados de um registro existente por completo dentro da API. Por comportamento padrão, caso o registro não exista para ser atualizado, é criado um novo registro.
    * DELETE - Utilizado para deletar um registro da API, tendo como base de busca um identificador único (Eg.: ID, CPF, ETC);
    * PATCH - Utilizado para aplicar modificações parciais em um registro existente, com um comportamento similar ao PUT porém somente em campos definidos;
    * OPTIONS - Método utilizado para um cliente poder descobrir quais requisições HTTP são permitidas dentro de um recurso ou servidor; 
    * HEAD - Utilizado para um cliente descobrir quais os cabeçalhos daquele determinado recurso ou servidor. Semelhante ao GET mas sem corpor (Eg. uma requisição HEAD retornaria algo como {"name", "CPF", ...} sem corpo de nome ou de CPF);
    * TRACE - É utilizado como método de diagnóstico, depuração, e análise de requições. Permite que o cliente veja como a requisição é recebida e modificada por servidores;. É um método pouco usado e geralmente desabilitado por motivos de segurança.

Ref.: [MDN](https://developer.mozilla.org/pt-BR/docs/Web/HTTP/Reference/Methods)

2. As principais semelhanças entre Redis e MongoDB são elas:
    * Ambos são sistemas de gerenciamento de banco de dados NoSQL, isso significa que são sistemas não relacionais;
    * Ambos foram lançados no ano de 2009;
    * Ambos podem ser escalados de forma horizontal, isso significa que é possível adicionar mais servidores ou nós ao sistema para distribuir a carga de trabalho, mais indicado para se trabalhar com alto volume de dados;
    * Ambos podem ser escalados de forma vertical, isso significa significa aumentar os recursos (como CPU, RAM ou armazenamento) de um único servidor para torná-lo mais potente (OBS: Destacando que o foco do MongoDB é escalabilidade horizontal, mas é possível ser feito vertical), indicado para se melhorar o desempenho de um único servidor;
    * Ambos são compatíveis com diversos sistemas operacionais e linguagens de programação;
    * Ambos os sistemas são de código aberto;<br><br>
3. As principais diferenças entre Redis e MongoB são elas:
    * MongoDB é projetado e mais adequado para manuzeio de grandes volumes de dados, enquanto o Redis possui como foco a velocidade de resposta;
    * Pelas características individuais de cada um, Redis está mais presente em aplicações como IoT, serviços de mensageria, serviços de streaming, jogos online, etc, enquanto MongoDB está mais presente em aplicações como desenvolvimento de serviços de IA, Redes Sociais, empresas que prestam serviçços financeiros, etc.
    * MongoDB armazena dados (estruturados, semiestruturados e não estruturados) como documentos no formato BSON, como um JSON mas em formato binário, já o Redis armazena os dados em memória, direto na RAM, em uma estrutura de chave-valor com diferentes tipos de dados, como Strings, Lists, Sets, etc.
    * MongoDB foi projetado para ser escalado horizontalmente, sem restrições, isso garante uma estrutura mais flexível e concistente, enquanto o Redis também ter essa disponibilidade, não possui um sistema de divisão de informações in built, o que cria a necessidade de configurações manuais partindo do usuário;
    * Apesar de ambos serem sistemas de alta disponibilidade, o MongoDB possui configurações de failover automáticas, isso significa que possuirá melhor desempenho em detectar falhas em nós primários, mas também os corrige quando necessário, já o Redis, possui disponibilidade para configuração do failover manualmente, justamente pelo seu propósito de uso;
    * O MongoDB possui como linguagem de consulta o MQL, com uma sintaxe próxima a do JSON, enquanto o Redis não possui linguagem de consulta, fazendo-se necessário o uso de comandos para se realizar pesquisas de dados;
    * MongoDB possui serviços pagos, com recursos de seguranças adicionais e suporte profissional;

Ref.: [IONOS DIGITAL GUIDE](https://www.ionos.com/pt-br/digitalguide/servidor/conhecimento/mongodb-vs-redis/)

4. O Spring Boot utiliza principalmente o JUnit como framework base para testes unitários.

**Funcionamento do JUnit no Spring Boot**
* O JUnit organiza os testes em métodos anotados.

* Algumas anotações comuns:

    * @Test → marca um método como caso de teste.

    * @BeforeEach → executa um método antes de cada teste (geralmente para inicialização).

    * @AfterEach → executa depois de cada teste (geralmente para limpeza).

    * @BeforeAll e @AfterAll → executam antes/depois de todos os testes da classe.

**Integração com o Spring**

* O Spring Boot oferece suporte adicional via Spring Test e Spring Boot Test, que facilitam carregar o contexto do Spring durante os testes.

* Exemplo:

    * @SpringBootTest → sobe o contexto completo da aplicação para testes de integração.

    * @WebMvcTest → sobe apenas o contexto da camada web (controllers, filtros etc.).

**Mockito**

* Para isolar dependências, o Spring Boot integra bem com o Mockito, permitindo criar mocks (objetos simulados), isso evita precisar chamar banco de dados ou serviços reais durante um teste unitário.

* Exemplo de uso:

    * @Mock → cria um mock de uma classe.

    * @InjectMocks → injeta os mocks em uma classe a ser testada.

    * when(...).thenReturn(...) → define o comportamento simulado.

#### Explicação do desenvolvimento

Para este sistema, foram desenvolvidos os packages abaixo



Onde:

Audit - Implementa auditoria de ações do usuário no sistema, registrando quem criou ou modificou registros;<br>
Controller - Responsável por representar os endpoints de cada entidade;<br>
DTO - Data Transfer Objects responsáveis pela transação de informações no sistema, atua como uma camada de segurança entre o cliente e o banco de dados;<br>
Models - Responsáveis pela modelagem do banco de dados e representação das entidades;<br>
Repository - Extenção de JPARepository, atua em conjunto com H2 no repositório local;<br>
Security - Implementa a autenticação e autorização com JWT, incluindo login, registro e proteção de endpoints.<br>

#### Dentro de cada package, temos o seguinte

**Audit > SpringSecurityAuditorAware** 

Classe responsável por buscar no SecurityContext o usuário autenticado no momento da operação, através de um único método do AuditorAware chamado getCurrentAuditor. Esse método funciona como auxiliador de operações de auditoria automatica e fornece associações de operações no banco a um usuário.

**Controller > AlunoController**

Classe responsável por expor endpoints REST da entidade Aluno.
Funciona como intermediária entre cliente e banco de dados, recebendo requisições HTTP e delegando operações ao AlunoRepository.

Principais responsabilidades:

POST /alunos: cria um novo aluno a partir de um AlunoRequestDTO e retorna um AlunoResponseDTO.

GET /alunos/{id}: busca aluno pelo ID, retornando os dados em DTO.

GET /alunos: lista todos os alunos cadastrados.

PUT /alunos/{id}: atualiza dados de um aluno existente.

DELETE /alunos/{id}: remove um aluno pelo ID.
👉 Resumo: Controlador CRUD de alunos, garantindo a conversão entre entidades e DTOs.

**Controller > DisciplinaController** 

Classe responsável por expor endpoints REST da entidade Disciplina.
Gerencia requisições relacionadas a disciplinas, operando junto ao DisciplinaRepository.

Principais responsabilidades:

POST /disciplinas: cria nova disciplina a partir de um DisciplinaRequestDTO.

GET /disciplinas/{id}: busca disciplina por ID.

GET /disciplinas: lista todas as disciplinas cadastradas.

PUT /disciplinas/{id}: atualiza nome e código de uma disciplina.

DELETE /disciplinas/{id}: remove disciplina pelo ID.
👉 Resumo: Controlador CRUD para gerenciar disciplinas.

**Controller > MatriculaController**

Classe responsável por expor endpoints REST da entidade Matrícula.
Opera sobre a relação entre Aluno e Turma, garantindo a persistência via MatriculaRepository.

Principais responsabilidades:

POST /matriculas: cria uma nova matrícula relacionando aluno e turma a partir de MatriculaRequestDTO.

PUT /matriculas/{id}/notas: atualiza notas de uma matrícula existente.

GET /matriculas/{id}: retorna detalhes de uma matrícula.

GET /matriculas: lista todas as matrículas cadastradas.

DELETE /matriculas/{id}: remove matrícula pelo ID.
👉 Resumo: Controlador que centraliza o gerenciamento de matrículas, incluindo atualização de notas.

**Controller > ProfessorController**

Classe responsável por expor endpoints REST da entidade Professor.
Funciona como interface entre requisições externas e o ProfessorRepository.

Principais responsabilidades:

POST /professores: cria novo professor a partir de um ProfessorRequestDTO.

GET /professores/{id}: busca professor pelo ID.

GET /professores: lista todos os professores.

PUT /professores/{id}: atualiza dados de um professor existente.

DELETE /professores/{id}: remove professor pelo ID.
👉 Resumo: Controlador CRUD de professores.

**Controller > TurmaController** 

Classe responsável por expor endpoints REST da entidade Turma.
Relaciona turmas com Professor e Disciplina, utilizando os repositórios correspondentes.

Principais responsabilidades:

POST /turmas: cria uma nova turma vinculada a professor e disciplina via TurmaRequestDTO.

GET /turmas/{id}: busca turma pelo ID.

GET /turmas: lista todas as turmas cadastradas.

DELETE /turmas/{id}: remove turma pelo ID.
👉 Resumo: Controlador CRUD de turmas, responsável por vincular professores e disciplinas.

**DTO > AlunoRequestDTO**

Classe responsável por receber dados de entrada do cliente para criação/atualização de um Aluno.
Possui método toEntity() que converte o DTO em uma instância da entidade Aluno.

**DTO > AlunoResponseDTO**

Classe responsável por retornar ao cliente os dados de um Aluno.
Recebe uma entidade Aluno no construtor e expõe seus atributos em formato seguro para resposta HTTP.

**DTO > AtualizarNotasRequestDTO**

Classe responsável por receber os dados de atualização de notas de uma matrícula.
Utilizada em operações de atualização parcial (nota1, nota2, nota3).

**DTO > DisciplinaRequestDTO**

Classe responsável por receber dados de entrada do cliente para criação/atualização de uma Disciplina.
Possui método toEntity() que converte o DTO em uma instância da entidade Disciplina.

**DTO > DisciplinaResponseDTO**

Classe responsável por retornar ao cliente os dados de uma Disciplina.
Recebe uma entidade Disciplina no construtor e expõe seus atributos para resposta HTTP.

**DTO > LoginRequestDTO**

Classe responsável por transportar credenciais de login (e-mail e senha) fornecidas pelo cliente.
Utilizada no processo de autenticação.

**DTO > MatriculaDetalheResponseDTO**

Classe responsável por detalhar informações de uma Matrícula, incluindo notas, média e resultado do aluno.
No construtor, realiza lógica de cálculo da média e definição do resultado (Aprovado, Reprovado ou Não foram lançadas todas as notas).

**DTO > MatriculaRequestDTO**

Classe responsável por receber dados do cliente para criação de uma Matrícula.
Possui método toEntity(Aluno, Turma) que cria uma instância de matrícula já associada a aluno e turma.

**DTO > MatriculaResponseDTO**

Classe responsável por retornar informações resumidas de uma Matrícula.
Expõe o nome do aluno e o resultado (aprovado, reprovado ou notas não lançadas).

**DTO > ProfessorRequestDTO**

Classe responsável por receber dados de entrada para criação/atualização de um Professor.
Possui método toEntity() que converte o DTO em uma instância da entidade Professor.

**DTO > ProfessorResponseDTO**

Classe responsável por retornar dados de um Professor ao cliente.
Recebe uma entidade Professor no construtor e expõe apenas atributos essenciais (id, nome, email).

**DTO > RegisterRequestDTO**

Classe responsável por transportar dados necessários para o registro de um novo Professor (nome, e-mail e senha).
Usada durante o processo de cadastro na autenticação.

**DTO > TurmaRequestDTO**

Classe responsável por receber dados de entrada para criação de uma Turma.
Possui método toEntity(Professor, Disciplina) que cria a entidade já associando professor e disciplina.

**DTO > TurmaResponseDTO**

Classe responsável por retornar os dados de uma Turma ao cliente.
Expõe atributos próprios e também informações agregadas de Disciplina, Professor e Matriculas.

**Models > Aluno**

Entidade que representa um Aluno no sistema acadêmico.
Contém informações pessoais (nome, CPF, email, telefone, endereço) e mantém relacionamento 1:N com Matricula.
Implementa auditoria automática com @CreatedBy, @CreatedDate, @LastModifiedBy e @LastModifiedDate, armazenando dados de criação e atualização.

**Models > Disciplina**

Entidade que representa uma Disciplina.
Contém informações básicas como nome e código identificador.
Mantém relacionamento 1:N com Turma, permitindo vincular várias turmas à mesma disciplina.
Possui campos de auditoria para rastrear criação e atualizações.

**Models > Matricula**

Entidade que representa a Matrícula de um aluno em uma turma.
Relaciona-se com Aluno (N:1) e com Turma (N:1).
Armazena notas (nota1, nota2, nota3, notaFinal) e possui método calcularMedia() para retornar a média aritmética simples das três primeiras notas. Caso alguma nota seja nula, retorna null.

**Models > Professor**

Entidade que representa um Professor no sistema.
Contém atributos de identificação (nome, email, senha) e mantém relacionamento 1:N com Turma.
Assim como outras entidades principais, possui auditoria automática de criação e atualização.

**Models > Turma**

Entidade que representa uma Turma.
Relaciona-se com Professor (N:1) e Disciplina (N:1), além de manter lista de Matricula (1:N).
Registra auditoria com informações de quem criou e atualizou, além das datas correspondentes.

**Repository > AlunoRepository/DisciplinaRepository/MatriculaRepository/TurmaRepository e ProfessorRepository<br>**

Todas as classes extendem JpaRepository, criando assim um banco de dados JPA em conexão com o H2. A única diferenciação, em ProfessorRepository, temos um método findByEmail(String email), utilizado no momento do login, para localização do email do professor.

**Security > AuthController**

Controlador REST responsável por gerenciar autenticação e registro de professores.

/auth/login: autentica o usuário com email e senha, gera e retorna um token JWT em caso de sucesso.

/auth/register: realiza o cadastro de novos professores, validando email único e armazenando a senha com criptografia (PasswordEncoder).

**Security > JwtAuthFilter**

Filtro customizado que intercepta todas as requisições (via OncePerRequestFilter) e valida tokens JWT no cabeçalho Authorization.

Extrai o token, recupera o usuário correspondente e insere a autenticação no SecurityContextHolder.

Caso o token seja inválido ou ausente, a requisição segue normalmente sem autenticação.

**Security > JwtUtils**

Classe utilitária para manipulação de tokens JWT.

gerarToken(String username): cria um token assinado com chave secreta e tempo de expiração configurado.

extrairUsername(String token): obtém o usuário associado ao token, mesmo que esteja expirado.

validarToken(String token, String username): confirma se o token é válido, pertence ao usuário e não expirou.

Implementa também a verificação de expiração via método estaExpirado().

**Security > ProfessorUserDetails**

Classe que implementa UserDetails, adaptando a entidade Professor ao modelo de autenticação do Spring Security.

Retorna credenciais (email, senha) e informações sobre a conta.

Atualmente não utiliza permissões (getAuthorities() retorna lista vazia).

Fornece acesso ao professor_id para uso em contexto de segurança.

**Security > ProfessorUserDetailsService**

Serviço que implementa UserDetailsService para buscar professores pelo email.

Utiliza ProfessorRepository para consultar o banco.

Caso o professor não seja encontrado, lança exceção UsernameNotFoundException.

Retorna instância de ProfessorUserDetails para integração com Spring Security.

**Security > SecurityConfig**

Classe de configuração principal do Spring Security.

Define a cadeia de filtros (SecurityFilterChain), aplicando o JwtAuthFilter antes do UsernamePasswordAuthenticationFilter.

Desabilita CSRF e configura regras de autorização: /auth/** é público, demais endpoints exigem autenticação.

Define beans para AuthenticationProvider, PasswordEncoder (BCrypt) e AuthenticationManager.
